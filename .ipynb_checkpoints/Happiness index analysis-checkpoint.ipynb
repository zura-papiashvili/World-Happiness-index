## მსოფლიო ქვეყნების ბედნიერების ინდექსი

**კონტექსტი**

მსოფლიოს ბედნიერების რეპორტი არის გლობალური ბედნიერების მდგომარეობის საყოველთაო გამოკითხვა. პირველად ეს რეპორტი გამოქვეყნდა 2012 წელს და 2016 წლამდე ახლდებოდა. მას შემდეგ კი სტაბილურად გამოიყენებოდა სხვადასხვა სფეროში: ეკონომიკა, ფსიქოლოგია, ნაციონალური სტატისტიკა, ჯანდაცვა, საყოველთაო დაზღვევა და სხვა. 



# **კონტენტი**

ბედნიერების ქულები იყენებს  *Gallup World Poll*-ის მონაცემებს. ქულები ეფუძვნება მთავარი ცხოვრებისეული საკითხების შეფასების კითხვებს, რომელებიც გამოკითხვაში იყო გამოყენებული. ეს კითხვები ცნობილია როგორც Cantril ladder

ბედნიერების ქულების  შემდეგი სვეტები ასასხავს 6 ფაქტორს რომლებიც განსაზღვრავენ ამ გამოკითხვის შედეგებს: 

- ეკონომიკა

- თავისუფლება
- კორუფციის არ არსებობა
- კეთილდღეობა

ჩამოთვლილი ფაქტორები განსაზღვრავენ თუ რამდენად ბედნიერი ხალხი ცხოვრობს თითოეულ ქვეყანაში, "დისტოპია"-ში მცხოვრებ ხალხთან შედარებით.

**რა არის დისტოპია?**

დისტოპია არის წარმოსახვითი ქვეყანა, რომელშიც მსოფლიოში ყველაზე უბედური ხალხი ცხოვრობს. თითოეული გამოკითხული ადამიანი სწორედ ასეთ ქვეყანასთან აკეთებს შედარებას სადაც იქნება ყველაზე დაბალი შემოსავალი, უმცირესი ცხოვრებისეული მოლოდინი, დაბალი კეთილდღეობა, მაღალი კორუფცია, თავისუფლების ნაკლებობა და უმცირესი სოციალური დახმარება. შედარება ხდება უტოპიის საწინააღმდეგო მდგომარეობასთან "დისტოპია"-სთან.

```python
# შემოვიტანოთ სუიციდის შემთვევების მონაცემები ქვეყნების და წლების მიხედვით:

suicide = pd.read_csv('data/share-deaths-suicide.csv')
suicide.head()
```

```python
# შევუცვალოთ სახელები სვეტებს:

suicide=suicide[['Entity','Year','Deaths - Self-harm - Sex: Both - Age: All Ages (Percent)']].rename(columns={'Entity':'Country','Deaths - Self-harm - Sex: Both - Age: All Ages (Percent)':'suicide'})
suicide.head()
```

```python
# დავტოვოთ მხოლოდ 2015 წლის მონაცემები და ვნახოთ  ინფო
suicide=suicide[suicide['Year']==2015]
suicide.info()
```

```python
# როგორც ჩანს მონაცემებში არ არის Nan-ები ამიტომ შეგვიძლია გავაერთიანოთ მონაცემები:

df = pd.merge(y_2015, suicide, on=['Country'])
```

```python
# წავშალოთ სვეტი Year
df.drop(columns=['Year'],inplace=True)
df.head()
```

```python
df.info()
```

```python
# ავაგოთ კორელაქციის რუკა:
sns.heatmap(df[['Happiness Score',
       'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual','suicide']].corr(),annot=True)
```

```python
# ვნახოთ უფრო ახლოდან თუ როგორ არის სუიციდიი და ბედნიერების ინდექსი ურთიერთდამოკიდებული
sns.jointplot(x='Happiness Score',y='suicide',data=df)
```

```python
#ასევე შევადაროთ ეკონომიკურ მდგომარეობას
sns.jointplot(x='Economy (GDP per Capita)',y='suicide',data=df)
```

**როგორც გრაფიკებიდან ჩანს უმნიშვნელოდ მაგრამ მაინც შეიმჩნევა კორელაცია სუიციდსა და  ეკომოიკურ მდომარეობას შორის.
რაც ნიშნავს რომ ეკონომიკურად განვითარებულ ქვეყნებში გამოკვლეულ პერიოდში პროცენტულად უფრო მეტი სუიციდის შემთხვევა იყო.**


```python
sns.relplot(data=df,x='Economy (GDP per Capita)',y='Happiness Score',hue='Region',size="suicide",sizes=(1, 200))
```

```python
df.sort_values("suicide",ascending=False).head(10)
```

```python
from sklearn.preprocessing import OneHotEncoder
# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')
```

```python
enc_df = pd.DataFrame(enc.fit_transform(df[['Region']]).toarray())
dum_df = pd.get_dummies(df, columns=["Region"])
```

```python
X=dum_df.drop(columns=["Country",'Happiness Rank','Trust (Government Corruption)','Generosity','Standard Error','Dystopia Residual','suicide'])
```

```python
y=dum_df['suicide']
```

```python
from sklearn.linear_model import LinearRegression

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)


regressor = LinearRegression()
regressor.fit(X_train, y_train)


```

```python
y_pred = regressor.predict(X_test)
```

```python
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
predictions
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
